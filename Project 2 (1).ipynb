```python
from sklearn.datasets import fetch_california_housing
```

```python
cal_housing = fetch_california_housing()
```

```python
from sklearn.model_selection import train_test_split
```

```python
print("Features: ", cal_housing.feature_names)
print("Target: ", cal_housing.target)
```

```python
for i in range(8):
    print(cal_housing.feature_names[i])
    plt.scatter(cal_housing.data[:,i], cal_housing.target)
    plt.show()
    plt.xlabel(cal_housing.feature_names[i])
    plt.ylabel('Price')
  
```

```python
X_train, X_test, y_train, y_test = train_test_split(cal_housing['data'], cal_housing['target'], random_state = 0)
```

```python
cal_dataframe = pd.DataFrame(X_train, columns = cal_housing.feature_names)
```

```python
grr = pd.plotting.scatter_matrix(cal_dataframe, c = y_train, figsize = (15,15), marker = 'o', hist_kwds = {'bins': 20
}, s = 60, alpha = 0.8 )
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
print(cal_housing['data'].shape)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
```

```python
print("X_test.shape: {}".format(X_test.shape))
print("y_test shape: {}".format(X_test.shape))
```

```python
lab_enc = preprocessing.LabelEncoder()
y_train_encoded = lab_enc.fit_transform(y_train)
```

```python
knn.fit(X_train, y_train_encoded)
```

```python
y_pred = knn.predict(X_test)
```

```python
print("Test set score:\n {}".format(y_pred))
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```
